```python
import requests
import pandas as pd
import numpy as np
```

```python
import lxml.html as lh
```

```python
url='http://www.coronaworldpulse.com/'
```

```python
page = requests.get(url)
```

```python
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
i=0
col=[]
for z in range(0,len(tr_elements)):
    for t in tr_elements[z]:
        i+=1
        name=t.text_content()
        print(i,name)
        col.append((name))
col
```
