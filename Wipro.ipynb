```python
# event filtering via event stream
def import_xes(x):
  event_log = pm4py.read_xes(x)
  event_stream = pm4py.convert_to_event_stream(event_log)
  
  permit_submitted = functools.filter_(lambda e: e['concept:name'] == 'Permit SUBMITTED by EMPLOYEE', event_stream)
  
  permit_permitted = functools.filter_(lambda e: e['concept:name'] == 'Permit FINAL_APPROVED by SUPERVISOR', event_stream)
  
  declaration_submitted = functools.filter_(lambda e: e['concept:name'] == 'Declaration SUBMITTED by EMPLOYEE', event_stream)
  
  declaration_permitted = functools.filter_(lambda e: e['concept:name'] == 'Declaration FINAL_APPROVED by SUPERVISOR', event_stream)

  request_payment = functools.filter_(lambda e: e['concept:name'] == 'Request Payment', event_stream)

  Payment_given = functools.filter_(lambda e: e['concept:name'] == 'Payment Handled', event_stream)


  
  
  Permit_submit_count = 0
  Permit_permited_count = 0
  Dec_submit_count = 0
  Dec_permited_count = 0
  request_payment_count = 0
  payment_given_count = 0
  
  for e in permit_submitted:
    Permit_submit_count+=1
    
  for e in permit_permitted:
    Permit_permited_count+=1

  for e in declaration_submitted:
    Dec_submit_count+=1

  for e in declaration_permitted:
    Dec_permited_count+=1  

  for e in request_payment:
    request_payment_count+=1  
  
  for e in Payment_given:
    payment_given_count+=1  

  list = [Permit_submit_count , Permit_permited_count , Dec_submit_count , Dec_permited_count , request_payment_count , payment_given_count ]
  return(list)

```

```python
#count = import_xes('PermitLog.xes_')
count = import_xes('DomesticDeclarations.xes_')
count2 = import_xes('InternationalDeclarations.xes_')
```
